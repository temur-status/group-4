Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract data
months = data["month_number"]
toothpaste_sales = data["toothpaste"]
# Add legend

# Create scatter plot
plt.scatter(months, toothpaste_sales, label="Toothpaste sales")
plt.legend(loc="upper left")
# Add labels
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")

# Add grid
plt.grid(True, linestyle='--')

# Show plot
plt.show()
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
months = data["month_number"]
facecream = data["facecream"]
facewash = data["facewash"]

# Bar width
bar_width = 0.4

# Positions for bars
r1 = np.arange(len(months))
r2 = r1 + bar_width

# Create bar chart
plt.bar(r1, facecream, width=bar_width, label="Face Cream")
plt.bar(r2, facewash, width=bar_width, label="Face Wash")

# Add labels
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")

# Add x-axis ticks
plt.xticks(r1 + bar_width / 2, months)
# Add grid
plt.grid(True, linestyle='--')
# Add legend
plt.legend()

# Show plot
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Define colors for each product
total_sales = data[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum()

colors = ['orange', 'lightblue', 'yellow', 'red', 'green', 'pink']  # red for bathing soap

# Plot Pie chart
plt.figure(figsize=(8,8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140, shadow=True, colors=colors)
plt.title('Total Units Sold Per Product in Last Year')
plt.legend(loc='upper left')
plt.show()
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
fig, axes = plt.subplots(2, 1, figsize=(10,8))

# Plot Bathing Soap sales
axes[0].plot(data['month_number'], data['bathingsoap'], marker='o', color='red')
axes[0].set_title('Bathing Soap Sales')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Units Sold')

# Plot Facewash sales
axes[1].plot(data['month_number'], data['facewash'], marker='s', color='blue')
axes[1].set_title('Facewash Sales')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Units Sold')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

The Stack plot should look like this:

```python
months = data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']

# Colors for each product
colors = ['orange', 'lightblue', 'yellow', 'red', 'green', 'pink']

# Plot stackplot
plt.figure(figsize=(12,6))
plt.stackplot(months, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer,
              labels=['Face Cream','Face Wash','Toothpaste','Bathing Soap','Shampoo','Moisturizer'],
              colors=colors)

# Add labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Sales of All Products (Stack Plot)')
plt.legend(loc='upper left')

# Show plot
plt.show()
```
