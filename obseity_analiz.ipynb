```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("./obseity.csv")
```

```python
gender_counts = data["Gender"].value_counts()

labels=[f"{gender} ({count})" for gender, count in zip(gender_counts.index, gender_counts)]

plt.pie(
        gender_counts,
        labels=labels,
        autopct="%1.1f%%",
        startangle=90,
        colors=["red","green"]
)

plt.title("Erkaklar va Ayollar taqsimoti")
plt.show()
```

```python
import matplotlib.pyplot as plt

mapping = {
    0: "O'ta ozg'in",
    1: "Ozg'in",
    2: "Yaxshi",
    3: "Biroz semiz",
    4: "Semiz",
    5: "O'ta Semiz"
}

index_counts = data["Index"].value_counts().sort_index()

colors = [
    "#8BC34A",
    "#CDDC39",
    "#FFEB3B",
    "#FFC107",
    "#FF5722",
    "#F44336"
]

plt.bar(index_counts.index, index_counts.values, color=colors)

for a, b in zip(index_counts.index, index_counts.values):
    plt.text(a, b+3, str(b), ha='center', fontweight='bold')

plt.xlabel("Index darajasi")
plt.ylabel("Odamlar soni")
plt.title("Index bo'yicha odamlar soni")
plt.xticks(index_counts.index, [mapping[a] for a in index_counts.index], rotation=16)

plt.show()
```

```python
import matplotlib.pyplot as plt

male_heights = data[data["Gender"] == "Male"]["Height"]
female_heights = data[data["Gender"] == "Female"]["Height"]

plt.boxplot(
    [male_heights, female_heights],
    tick_labels=["Erkaklar", "Ayollar"],
    patch_artist=True
)

plt.title("Erkaklar va ayollarning bo'y nisbatlari")
plt.xlabel("Gender")
plt.ylabel("Bo'y")
plt.show()
```

```python
import matplotlib.pyplot as plt

male_weights = data[data["Gender"] == "Male"]["Weight"]
female_weights = data[data["Gender"] == "Female"]["Weight"]

plt.boxplot(
    [male_weights, female_weights],
    tick_labels=["Erkaklar", "Ayollar"],
    patch_artist=True
)

plt.title("Erkaklar va ayollarning vazn nisbatlari")
plt.xlabel("Gender")
plt.ylabel("Vazn")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(data["Height"], bins=20, edgecolor="black")
plt.title("Bo'y balandligi taqsimoti")
plt.xlabel("Bo'y balandligi")
plt.ylabel("Odamlar soni")
plt.show()
```

```python
import matplotlib.pyplot as plt

data["BMI"] = data["Weight"] / (data["Height"]/100)**2

plt.hist(data["BMI"], bins=20, edgecolor="black")
plt.title("BMI taqsimoti")
plt.xlabel("BMI qiymati")
plt.ylabel("Odamlar soni")
plt.show()
```
